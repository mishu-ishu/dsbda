```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display the first few rows of the dataset to understand its structure
print(titanic.head())

# Plotting the histogram for the 'fare' column to see how ticket prices are distributed
plt.figure(figsize=(10,6))
sns.histplot(titanic['fare'], kde=True, bins=30)  # kde=True adds a Kernel Density Estimate (smooth curve)
plt.title('Distribution of Ticket Prices (Fare) for Titanic Passengers')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
